# Fitting Our ARIMA Model

```python
#Checking for trends
rolmean = pd.DataFrame(data_cleaned_df.Total_Accident.rolling(window = 8, center = False).mean())
rolstd = pd.DataFrame(data_cleaned_df.Total_Accident.rolling(window = 8, center = False).std())
```

```python
rolmean.head(15)
```

```python
#plotting rolling and data together
fig = plt.figure(figsize=(15,8))
data_cleaned_df.Total_Accident.plot(color='blue',label='Original')
rolmean.Total_Accident.plot(color='red', label='Rolling Mean')
rolstd.Total_Accident.plot(color='black', label = 'Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.show(block=False)
```

```python
#perform dickey fuller to see if my data is stationary
def check_stationarity(df):
    index =np.array(df.index)
    passe= np.array(df['Total_Accident'])
    ts = pd.Series(passe, index=index)
    from statsmodels.tsa.stattools import adfuller
    dftest = adfuller(ts)
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    return print(dfoutput)
```

```python
check_stationarity(data_cleaned_df)
```

```python
#Split my data into 2 groups: one for training and another for fitting
split_size = round(.8* len(data_cleaned_df))
split_size
```

```python
training = data_cleaned_df[:split_size]
test = data_cleaned_df[split_size:]
```

```python
#Out-of-sample validation of an ARIMA model

# Forecast
fc, se, conf = model_fit.forecast(len(test), alpha=0.05) # 95% conf
    # Make as pandas series
fc_series = pd.Series(fc, index=test.index)
lower_series = pd.Series(conf[:, 0], index=test.index)
upper_series = pd.Series(conf[:, 1], index=test.index)
    # Plot
plt.figure(figsize=(16,10), dpi=100)
training.Total_Accident.plot(label='training')
test.Total_Accident.plot(label='actual')
plt.plot(fc_series, label='forecast')
plt.fill_between(lower_series.index, lower_series, upper_series,
color='k', alpha=.15)
plt.title('Forecast vs Actuals')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```
